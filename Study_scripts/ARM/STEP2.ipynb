```python
import pandas as pd
from csv import writer
import os
import csv
```

```python

### put the longest row as first

p='/Projects'
dirlist = [item for item in os.listdir(p) if os.path.isdir(os.path.join(p, item))]
# print (dirlist)
lenth = len(dirlist)
for index in range(0, lenth):
    project = dirlist[index]
    print(project)
    ProjectClone= p + "/" + project

    try:

# Open the CSV file
        with open(ProjectClone+'/(STEP1)_FINAL-IaCFilesCommits.csv', newline='') as csvfile: 
            reader = csv.reader(csvfile)
            

    # Find the lengthiest row
            max_len = 0
            max_row = []
            for row in reader:
                    if len(row) > max_len:
                        max_len = len(row)
                        max_row = row
            print(max_len)
        with open(ProjectClone+'/(STEP1)_FINAL-IaCFilesCommits.csv', 'r+') as csvfile:
            reader = csv.reader(csvfile)
            rows = [r for r in reader]

   
            max_row_index = rows.index(max_row)

    
            rows.insert(0, rows.pop(max_row_index))
            
    # Write the updated rows back to the CSV file
            csvfile.seek(0)
            writer = csv.writer(csvfile)
            writer.writerows(rows)
    except FileNotFoundError:
            # Handle missing commits
            pass    
```

```python


p='/Projects'
dirlist = [item for item in os.listdir(p) if os.path.isdir(os.path.join(p, item))]
# print (dirlist)
lenth = len(dirlist)
for index in range(0, lenth):
    project = dirlist[index]
    print(project)
    ProjectClone = p + "/" + project

    try:
          
          # Load the CSV file into a pandas dataframe
          df = pd.read_csv(ProjectClone+'/(STEP1)_FINAL-IaCFilesCommits.csv')

          # Calculate the number of rows to keep (80% of total rows)
          rows_to_keep = int(len(df) * 0.9)
          print(rows_to_keep)

          # Select the first 80% of the rows
          df_keep = df.iloc[:rows_to_keep]

          # Select the last 20% of the rows
          df_test = df.iloc[rows_to_keep:]

          # Save the result to a new CSV file
          df_keep.to_csv(ProjectClone+'/(STEP2)_90%_Of_Commits.csv', index=False)
          df_test.to_csv(ProjectClone+ '/(STEP7)_10%_Of_Commits.csv', index=False)




    except FileNotFoundError:
            # Handle missing commits
            pass    

```
